# Stage #6 — Conclusion and Recommendation

# 1. Conclusion

Berdasarkan hasil analisis statistik dan regresi pada dataset Diamonds, diperoleh beberapa kesimpulan sebagai berikut:

## Hasil Statistical Test

Independent sample t-test dilakukan untuk membandingkan rata-rata harga diamond dengan kategori cut Premium dan Ideal.

Hasil pengujian menunjukkan bahwa:

* nilai p-value < 0.05,
* dan nilai absolut t-statistic lebih besar dibanding t-critical.

Dengan demikian, H0 ditolak, sehingga dapat disimpulkan bahwa terdapat perbedaan rata-rata harga diamond yang signifikan antara kategori Premium dan Ideal.

---

## Hasil Analisis Regresi

Model regresi menunjukkan bahwa variabel:

* carat,
* cut,
* color,
* dan clarity

memiliki pengaruh signifikan terhadap harga diamond.

Variabel `log_carat` merupakan prediktor paling kuat terhadap harga diamond. Hal ini menunjukkan bahwa semakin besar ukuran diamond, maka harga diamond cenderung meningkat secara signifikan.

Selain itu:

* kualitas cut yang lebih baik,
* warna yang lebih baik,
* dan clarity yang lebih tinggi

juga meningkatkan harga diamond.

Interaction term antara `log_carat` dan `clarity` menunjukkan bahwa pengaruh carat terhadap harga berbeda pada setiap tingkat clarity diamond.

---

## Hasil Evaluasi Model

Model regresi akhir memiliki performa yang sangat baik dengan:

* Adjusted R-squared sebesar 0.983,
* dan rata-rata Cross Validation R-squared sebesar 0.9828.

Hal ini menunjukkan bahwa model mampu menjelaskan sebagian besar variasi harga diamond dan memiliki kemampuan generalisasi yang baik terhadap data baru.

Transformasi logaritma juga berhasil:

* mengurangi skewness,
* memperbaiki distribusi residual,
* dan meningkatkan performa model regresi.

---

# 2. Recommendation for Business

Berdasarkan hasil analisis, beberapa rekomendasi bisnis yang dapat diberikan adalah:

1. Fokus pada diamond dengan ukuran carat yang lebih besar karena ukuran diamond memiliki pengaruh paling besar terhadap harga.

2. Kualitas clarity dan cut perlu diperhatikan karena kualitas yang lebih baik dapat meningkatkan nilai jual diamond secara signifikan.

3. Strategi pricing dapat mempertimbangkan kombinasi antara carat dan clarity karena pengaruh carat terhadap harga berbeda pada setiap kategori clarity.

4. Perusahaan dapat menggunakan model regresi ini sebagai dasar untuk:

   * estimasi harga diamond,
   * penentuan harga jual,
   * dan evaluasi nilai inventory diamond.

Namun demikian, model masih memiliki keterbatasan karena belum mempertimbangkan faktor lain seperti:

* sertifikasi diamond,
* brand,
* kondisi pasar,
* dan asal diamond.

---
